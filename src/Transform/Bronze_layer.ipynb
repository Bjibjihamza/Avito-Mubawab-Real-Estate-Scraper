```python
from pyspark.sql import functions as F

storage_account = "strealestatehamza"
container = "realestate"

raw_base_path = f"abfss://{container}@{storage_account}.dfs.core.windows.net/raw"

# Options de lecture des CSV RAW (comme à l’ingest)
common_options = {
    "header": "true",
    "inferSchema": "false",  # on garde tout en string au début
}
```

```python
# Avito : ventes + locations
avito_ventes_raw_path    = f"{raw_base_path}/avito/ventes/*/*/*/*"
avito_locations_raw_path = f"{raw_base_path}/avito/locations/*/*/*/*"

avito_ventes_df_raw = spark.read.options(**common_options).csv(avito_ventes_raw_path)
avito_locations_df_raw = spark.read.options(**common_options).csv(avito_locations_raw_path)

print("Avito ventes RAW:", avito_ventes_df_raw.count())
print("Avito locations RAW:", avito_locations_df_raw.count())

# Mubawab : ventes + locations
mubawab_ventes_raw_path    = f"{raw_base_path}/mubawab/ventes/*/*/*/*"
mubawab_locations_raw_path = f"{raw_base_path}/mubawab/locations/*/*/*/*"

mubawab_ventes_df_raw = spark.read.options(**common_options).csv(mubawab_ventes_raw_path)
mubawab_locations_df_raw = spark.read.options(**common_options).csv(mubawab_locations_raw_path)

print("Mubawab ventes RAW:", mubawab_ventes_df_raw.count())
print("Mubawab locations RAW:", mubawab_locations_df_raw.count())
```
